```python
#Recursion
def num(n):
    if (n==0):  #Base case
        return
    print(n)
    num(n-1)
num(5)
```

```python
def fact(n):
    if (n==1 or n==0):
        return 1
    return fact(n-1)*n
print(fact(5))
```

```python
def natural(n):
    if (n==0):
        return 0
    return natural(n-1)+n
sum=natural(5)
print(sum)

    
```

```python
def recur(list,index=0):
    if (index==len(list)):
        return
    print(list[index])
    recur(list,index+1)

sports=["Cricket","Football","baseball","hockey","batminton"]
recur(sports)
```
